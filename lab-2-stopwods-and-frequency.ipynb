```python
data = pd.read_csv("../input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv")
```

```python
data_of_5 = data.head(5)
print(data_of_5)
```

```python
import nltk
import re
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stopwords_list = stopwords.words("english")
```

```python
curpos = []
for i in range(len(data_of_5)):
    review = re.sub('[^a-zA-z0-9]', ' ',data_of_5['review'][i])
    review = review.lower()
    review = review.split()
    review = [ word for word in review if not word in stopwords_list]
    review = ' '.join(review)
    curpos.append(review)
    
```

```python
curpos
```

```python
from nltk import ngrams , FreqDist
```

```python
wordfreq = []
for w in curpos:
    wordfreq.append(curpos.count(w))
    print(curpos,wordfreq)
```
